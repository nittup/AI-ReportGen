```python
from trafilatura import fetch_url, extract

# url = "https://edu.iflytek.com/about-us/news/company-news/2091"
    
for i, url in enumerate(urls, start=1):
    downloaded = fetch_url(url)
    if downloaded:
        content = extract(downloaded, include_comments=False, output_format="txt")
        print(f"Content from {url} saved")
        # 写入到/save目录
        with open(f"save/{i}.txt", "a", encoding="utf-8") as f:
            f.write(f"Content from {url}:\n{content}\n\n")       
    else:
        print(f"Failed to download content from {url}")
```
